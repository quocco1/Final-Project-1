## 3.3. Detect & remove outlier using IQR

```python
def detect_outliers_iqr(df, column):
    """
    Detect outliers using the IQR method.
    Outliers = values < Q1 - 1.5*IQR  or > Q3 + 1.5*IQR
    """
    if column not in df.columns:
        raise KeyError(f"Column '{column}' not found in DataFrame.")

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = max(Q1 - 1.5 * IQR, 1)
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outliers, low, high = detect_outliers_iqr(df, "Quantity")

print("Lower bound:", low)
print("Upper bound:", high)
print("Số outlier:", len(outliers))
```

```python
df_clean=df.drop(outliers.index)
```

# 4. Create Recency, Frequency, Monetary Column

```python
df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"])

# Build RFM
reference_date = df_clean["InvoiceDate"].max() + pd.Timedelta(days=1)

rfm = df_clean.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (reference_date - x.max()).days,  # Recency
    "CustomerID": "count",                                    # Frequency
    "TotalAmount": "sum"                                       # Monetary
})

rfm.columns = ["R", "F", "M"]

# Min-Max Scaling
scaler = MinMaxScaler()
rfm_scaled = pd.DataFrame(
    scaler.fit_transform(rfm),
    columns=["R_scaled", "F_scaled", "M_scaled"],
    index=rfm.index
)

# Combine original + scaled
rfm_final = pd.concat([rfm, rfm_scaled], axis=1)

print(rfm_final)
```

## 4.1. Calculate Weighted RFM Score

```python
rfm_final['RFM_Score']= rfm_final['R_scaled']*0.3 + rfm_final['F_scaled']*0.5 + rfm_final['M_scaled']*0.2
rfm_final
```

## 4.2. Assign VIP Labels

```python
import numpy as np

rfm_final['VIP'] = np.where(rfm_final['RFM_Score'] > 0.75, 1, 0)
rfm_final
```
